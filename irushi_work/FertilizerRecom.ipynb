```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("fertilizer_data.csv")

# Display the first few rows
df.head()
```

```python
def get_fertilizer_recommendations(plant, deficiency):
    # Convert inputs to lowercase for case-insensitive matching
    plant = plant.lower()
    deficiency = deficiency.lower()

    # Filter the DataFrame based on the input plant and deficiency
    result = df[(df["Plant"].str.lower() == plant) &
                (df["Deficiency"].str.lower() == deficiency)]

    if not result.empty:
        precautions = result.iloc[0]["Precautions"]
        fertilizers = result.iloc[0]["Fertilizers"]

        print(f"üå± **{plant.capitalize()} - {deficiency.capitalize()} Deficiency**")
        print("\n‚úÖ **Precautions:**")
        for p in precautions.split(";"):
            print(f"- {p.strip()}")

        print("\nüåø **Recommended Fertilizers:**")
        for f in fertilizers.split(";"):
            print(f"- {f.strip()}")
    else:
        print("‚ùå No data found for the given plant and deficiency.")

# Example Usage
get_fertilizer_recommendations("banana", "potassium")
```
